# Subset and Clean for Final/Full Version
## Workflow
* subset and rename columns
    * Rename: "fips", "state", "county", "perc_no_hs_{year}", "perc_col_grad_{year}"
* format FIPS code to 5-digit string
* extract national numbers and adjust all county percentaged to be % points above/below the national average for a given year
* separate county and state data into two separate Dataframes (and eventually CSVs)
* pivot states data to wide format for two separate time series

```python
states_filter = (raw['FIPS Code'] % 100 == 0) & raw['2003 Rural-urban Continuum Code'].isna()
```

```python
subset = raw[[
    "FIPS Code", 
    "State", 
    "Area name", 
    'Percent of adults with less than a high school diploma, 1970',
    'Percent of adults with less than a high school diploma, 1980',
    'Percent of adults with less than a high school diploma, 1990',
    'Percent of adults with less than a high school diploma, 2000',
    "Percent of adults with less than a high school diploma, 2015-19",
    'Percent of adults completing four years of college or higher, 1970',
    'Percent of adults completing four years of college or higher, 1980',
    "Percent of adults with a bachelor's degree or higher, 1990",
    "Percent of adults with a bachelor's degree or higher, 2000",
    "Percent of adults with a bachelor's degree or higher, 2015-19"
    ]].rename(columns = {
    "FIPS Code" : "fips", 
    "State" : "state", 
    "Area name" : "county",
    'Percent of adults with less than a high school diploma, 1970' : "no_hs_1970",
    'Percent of adults with less than a high school diploma, 1980' : "no_hs_1980",
    'Percent of adults with less than a high school diploma, 1990' : "no_hs_1990",
    'Percent of adults with less than a high school diploma, 2000' : "no_hs_2000",
    "Percent of adults with less than a high school diploma, 2015-19" : "no_hs_2019",
    'Percent of adults completing four years of college or higher, 1970' : "col_1970",
    'Percent of adults completing four years of college or higher, 1980' : "col_1980",
    "Percent of adults with a bachelor's degree or higher, 1990" : "col_1990",
    "Percent of adults with a bachelor's degree or higher, 2000" : "col_2000",
    "Percent of adults with a bachelor's degree or higher, 2015-19" : "col_2019"
})
subset.head()
```

```python
national_stats = subset.loc[0,:].to_dict()
national_stats
```

```python
# format fips int -> 5-digit str
subset.fips = subset.fips.map(lambda x: ("00000" + str(x))[-5:])
```

```python
# adjust percentages to be deviations from the national average
for metric in subset.columns[3:]:
    subset[metric] = subset[metric] - national_stats[metric]
```

```python
states = subset[states_filter & (subset.state != 'US')].copy()
print(states.shape)
counties = subset[~states_filter]
print(counties.shape)
```

```python
# melt down states data to friendly shape

# first for data on lacking high school ed
states_no_hs = states.melt(
    id_vars = ['fips', 'state'],
    value_vars = ['no_hs_1970', 'no_hs_1980', 'no_hs_1990', 'no_hs_2000', 'no_hs_2019'],
    var_name = 'year',
    value_name = 'no_hs'
)

states_no_hs['year'] = states_no_hs.year.map(lambda x: int(x[-4:]))

states_no_hs.head()
```

```python
# then for college grad
states_col = states.melt(
    id_vars = ['fips', 'state'],
    value_vars = ['col_1970', 'col_1980', 'col_1990', 'col_2000', 'col_2019'],
    var_name = 'year',
    value_name = 'col'
)

states_col['year'] = states_col.year.map(lambda x: int(x[-4:]))

states_col.head()
```

```python
# join the melted state-level data
states_long = states_no_hs.merge(
    states_col,
    on = ['fips','state','year'],
    how = 'outer'
)
print(states_long.shape)
states_long.head(5)
```

```python
sns.lineplot(
    data=states_long,
    x='year',
    y='no_hs',
    hue='state'
)

plt.legend([],[], frameon=False)
```

States have become more similar respective to the proportion of adults that have a high school diploma...

```python
sns.lineplot(
    data=states_long,
    x='year',
    y='col',
    hue='state'
)

plt.legend([],[], frameon=False)
```

But they have become more different respective of the proportion of adults with a college degree.
